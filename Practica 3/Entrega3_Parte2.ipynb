Ejercicio 2.3

La técnica seguida por el algoritmo j48 ofrece mejores resultados que la aplicada en el apartado anterior con el algoritmo id3 debido a que en dicho caso la discretización se realiza "a priori". Es decir, que además de no poder apoyarse en los valores de la clase de destino para la clasificación, este atributo tan solo podrá ser utilizado una vez para la clasificación.
Por contra, el algoritmo j48 realiza una técnica que aplica particionamiento binario y recursivo utilizando la Ganancia de Información como medida de bondad. La técnica de particionamiento recursivo de atributos numéricos permite una clasificación más eficiente ya que tan solo se produce cuando se maximiza la ganancia de información para dicha partición binaria.

Ejercicio 2.4

```python
#Procedemos a implementar la discretización
atri = [tabla.columns[i] for i in [1,2,15]]
atri
```

```python
for atributo in atri:
    b = []
    for pos in finditer(atributo+' '+'<= ', content_1):
        print (atributo+' '+'<= ', pos.start(), pos.end())
        if content_1[pos.end():].find(':') < content_1[pos.end():].find('\n'):
            segmento = content_1[pos.end():].find(':')
        else:
            segmento = content_1[pos.end():].find('\n')
            
        b.append(float(content_1[pos.end():pos.end()+segmento]))
        print(b[-1])
        b.sort()
    tabla[atributo] = [chr(65+i) for i in np.digitize(tabla[atributo], bins=b)]
```
